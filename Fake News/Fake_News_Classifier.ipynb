```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
%matplotlib inline
```

```python
df = pd.read_csv("train.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.fillna("Unavailable", inplace=True)
```

```python
df
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
X = df.drop(columns=["label"])
y = df["label"]
```

```python
X.shape, y.shape
```

```python
messages = X.copy()
```

```python
messages["title"][0]
```

```python
nltk.download("stopwords")
```

```python
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
corpus = []
for i in range(len(messages)):
    review = re.sub("[^a-zA-Z]"," ",messages["title"][i])
    review = review.lower().split()
    review = [lem.lemmatize(word) for word in review if not word in stopwords.words("english")]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
voc_size = 5000
```

```python
one_hot_rep = [one_hot(words,voc_size) for words in corpus]
```

```python
one_hot_rep[1]
```

```python
sent_length = 20
embedding_doc = pad_sequences(one_hot_rep,padding="post",maxlen=sent_length)
print(embedding_doc)
```

```python
embedding_doc[1]
```

```python
embedding_vector = 40
model = Sequential()
model.add(Embedding(voc_size, embedding_vector, input_length=sent_length))
model.add(Dropout(0.3))
model.add(LSTM(100))
model.add(Dropout(0.3))
model.add(Dense(1,activation="sigmoid"))
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
print(model.summary())
```

```python
X_final=np.array(embedding_doc)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=0)
```

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred=np.where(y_pred > 0.6, 1,0)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
